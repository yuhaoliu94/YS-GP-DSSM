```python
import os
import pickle
import numpy as np
from matplotlib import pyplot as plt

os.chdir("../../")

from src.gpdssm.utils import get_random_state
```

```python
Dx = 2
Dy = 20
T = 10000
C = 3
init_num = 1
X = [[0.0, 0.0]]
Y = []

random_state = get_random_state()
A = random_state.normal(0, 1, (Dy, Dx))
B = random_state.beta(1, 1, (Dy, Dx))
Z = np.zeros(T)
print(A)
print(B)
```

```python
def generate_z_process(T):
    cut_point = np.array([2,6,10,12,18,19]) * 500 #np.sort(random_state.choice(np.arange(T), 4, replace=False))
    Z[:cut_point[0]] = 0
    Z[cut_point[0]:cut_point[1]] = 1
    Z[cut_point[1]:cut_point[2]] = 2
    Z[cut_point[2]:cut_point[3]] = 1
    Z[cut_point[3]:cut_point[4]] = 2
    Z[cut_point[4]:cut_point[5]] = 1
    Z[cut_point[-1]:] = 0
```

```python
generate_z_process(T)
Z = np.array(Z)
```

```python
plt.plot(Z)
```

```python
def x_process_1(x):
    x_next_1 = 0.9 * x[0] + 0.5 * np.sin(x[1]) + random_state.normal(0., 1.0e-2)
    x_next_2 = 0.5 * np.cos(x[0]) + 0.9 * x[1] + random_state.normal(0., 1.0e-2)
    return [x_next_1, x_next_2]

def generate_x_process_1(T):
    x_next = x_process_1(X[-1])
    X.append(x_next)
```

```python
def x_process_2(x):
    x_next_1 = 0.9 * x[0] - 0.5 * np.sin(50 * x[0] * x[1]) + random_state.normal(0., 1.0e-2)
    x_next_2 = 0.5 * np.cos(50 * x[0] * x[1]) - 0.9 * x[1] + random_state.normal(0., 1.0e-2)
    return [x_next_1, x_next_2]

def generate_x_process_2(T):
    for t in range(T):
        x_next = x_process_2(X[-1])
        X.append(x_next)
```

```python
def x_process_3(x):
    x_next_1 = 0.9 * x[0] + 0.5 * np.sin(3 * x[1]) + random_state.normal(0., 1.0e-2)
    x_next_2 = 0.5 * np.sin(4 * x[0] + 1) + 0.9 * x[1] + random_state.normal(0., 1.0e-2)
    return [x_next_1, x_next_2]

def generate_x_process_3(T):
    for t in range(T):
        x_next = x_process_3(X[-1])
        X.append(x_next)
```

```python
def y_process(x):
    y = 0.5 * np.dot(A, x) + np.sin(np.dot(B, x)) + np.random.normal(0., 0.01, Dy)
    return y

def generate_y_process(T):
    for t in range(T):
        y_next = y_process(X[t+1])
        Y.append(y_next)
```

```python
generate_x = [generate_x_process_1, generate_x_process_2, generate_x_process_3]
```

```python
for i in Z:
    generate_x[int(i)](1)
X = np.array(X)
```
