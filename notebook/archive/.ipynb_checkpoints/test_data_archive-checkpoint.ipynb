```python
import os
import pickle
import numpy as np
from matplotlib import pyplot as plt

os.chdir("../")

from src.gpdssm.utils import get_random_state
```

```python
Dx = 2
Dy = 10
T = 10000
C = 3
init_num = 5
X = [[0.0, 0.0]]
Y = []

random_state = get_random_state()
A = random_state.normal(0, 1, (Dy, Dx))
B = random_state.beta(1, 1, (Dy, Dx))
rho = [random_state.beta(5, 1)] * 3
Z = [random_state.choice(np.arange(C))]
n = np.zeros(3)
n[Z[0]] = init_num
N = np.zeros(3)
N[Z[0]] = 1
print(A)
print(B)
print(rho)
```

```python
def get_prob_matrix(rho, n, N):
    matrix = np.zeros((C, C))
    for i in range(C):
        matrix[i, i] = n[i] / (rho[i] + n[i])
    matrix[0, 1] = 1 - matrix[0, 0]
    matrix[C-1, C-2] = 1 - matrix[C-1, C-1]
    for i in range(1, C-1):
        _sum = (rho[i+1] + N[i+1]) + (rho[i-1] + N[i-1])
        matrix[i, i+1] = (1 - matrix[i, i]) * (rho[i+1] + N[i+1]) / _sum
        matrix[i, i-1] = (1 - matrix[i, i]) * (rho[i-1] + N[i-1]) / _sum
    return matrix
```

```python
def generate_z_process(T):
    for i in range(T):
        matrix = get_prob_matrix(rho, n, N)
        z_cur = Z[-1]
        prob = matrix[z_cur, :]
        z_next = random_state.choice(np.arange(C), p=prob)
        if z_cur == z_next:
            n[z_cur] += 1
        else:
            n[z_cur] = 0
            n[z_next] = init_num
        N[z_next] += 1
        
        Z.append(z_next)
```

```python
generate_z_process(T)
Z = np.array(Z)
```

```python
def x_process_2(x):
    x_next_1 = 0.9 * x[0] - 0.5 * np.sin(50 * x[0] * x[1]) + random_state.normal(0., 1.0e-2)
    x_next_2 = 0.5 * np.cos(50 * x[0] * x[1]) - 0.9 * x[1] + random_state.normal(0., 1.0e-2)
    return [x_next_1, x_next_2]

def generate_x_process_2(T):
    for t in range(T):
        x_next = x_process_2(X[-1])
        X.append(x_next)
```

```python
def x_process_3(x):
    x_next_1 = 0.9 * x[0] + 0.5 * np.sin(3 * x[1]) + random_state.normal(0., 1.0e-2)
    x_next_2 = 0.5 * np.sin(4 * x[0] + 1) + 0.9 * x[1] + random_state.normal(0., 1.0e-2)
    return [x_next_1, x_next_2]

def generate_x_process_3(T):
    for t in range(T):
        x_next = x_process_3(X[-1])
        X.append(x_next)
```

```python
def y_process(x):
    y = 0.5 * np.dot(A, x) + np.sin(np.dot(B, x)) + np.random.normal(0., 0.01, 10)
    return y

def generate_y_process(T):
    for t in range(T):
        y_next = y_process(X[t+1])
        Y.append(y_next)
```
